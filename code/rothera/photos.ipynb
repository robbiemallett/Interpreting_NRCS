```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import string
from matplotlib import gridspec
```

```python
letters = string.ascii_lowercase

photos_dir = '../../photo_fig'

fnames = {'30/8 ':f'{photos_dir}/PXL_20230830_184727318.jpg',
          '30/8':f'{photos_dir}/PXL_20230830_150624816.jpg',
          '31/8':f'{photos_dir}/PXL_20230831_133046971.jpg',
          '04/9':f'{photos_dir}/PXL_20230904_192053605.jpg',
          '10/9':f'{photos_dir}/PXL_20230910_154712485.jpg',
          '15/9':f'{photos_dir}/15th_Sep_Vishnu.jpg',
          '20/9':f'{photos_dir}/20th_Sep_Robbie.jpg',
          '23/9':f'{photos_dir}/PXL_20230923_141818231.jpg',
          
         }
          

fig, axs = plt.subplots(4,2,figsize=(6,8))

for counter, (title, fname, ax, letter) in enumerate(zip(fnames.keys(),fnames.values(),
                                                         axs.reshape(-1), letters)):

    i = Image.open(fname)
    iar = np.array(i)
    
    if counter in [5]:    iar = iar[:-600]
    
    ax.imshow(iar)
    
    ax.set_title(title,pad=-2)
    

    t = ax.annotate(f'({letter})', fontsize='x-large', xy=(0.02,0.98),va='top',xycoords='axes fraction')
    
    t.set_bbox(dict(facecolor='white', alpha=0.5, edgecolor='none'))
    
    ax.set_yticks([])
    ax.set_xticks([])
    
plt.subplots_adjust(wspace=0.05)

plt.savefig('fieldphotos.jpg',dpi=700,bbox_inches='tight')
plt.savefig('/home/robbie/Dropbox/Apps/Overleaf/NRCS/figures/fieldphotos.png',dpi=700,bbox_inches='tight')
```
